## Time sampling

```python
df_tesla.head()
```

```python
# rule = A means annual , M is monthly
# it shows mininum of every year
df_tesla.resample(rule='M').min()
```

```python
df_tesla.resample(rule='M').max()
```

```python
df_tesla.resample(rule='W').max()['Open'].plot()
```

```python
df_tesla['Open'].resample(rule='BA').mean().plot(kind='bar')
```

## Rolling

```python
# rolling window, moving mean and window size is 10
df_tesla['High'].rolling(10).mean().head(20)
```

```python
df_tesla['Open: 30 days rolling'] = df_tesla['Open'].rolling(30).mean()
```

```python
plt.figure(figsize=(15,6))
sns.lineplot(df_tesla[['Open','Open: 30 days rolling']])
plt.show()
```

```python
# library to import time series data
from pandas_datareader import data as pdr
# yahoo finance data
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
yf.pdr_override()
```

```python
df = pdr.get_data_yahoo('MSFT',start='2017-01-01',end='2023-11-26')
```

```python
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(15,6))
sns.lineplot(df[['Open']])
plt.show()
```

```python
df.loc['2017-01-01':'2022-12-31'].astype(np.int64).max()
```

```python
open_series = df['Open'].sort_values()
print(f"Highest stock price: {open_series[-1]} USD\nDate: {open_series.index[-1]}")
```

```python
print(f"Lowest stock price: {open_series[0]} USD\nDate: {open_series.index[0]}")
```
